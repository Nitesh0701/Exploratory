```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
data=pd.read_csv('BostonHousing.csv')
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
#filling the null fields with the mean of that column.
col=['crim','zn','indus','chas','age','lstat']
for c in col:
  data[c].fillna(data[c].mean(),inplace=True)
```

```python
data.info()
```

```python
# checking the correlation of our target column that is ‘TAX’ with other columns.
data.corr()['tax'].sort_values(ascending=False)
```

```python
 from sklearn.preprocessing import StandardScaler
 X =data.drop('tax',axis=1)
 y=data['tax']

 scaler=StandardScaler()
 X=scaler.fit_transform(X)
 X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

rfc=RandomForestRegressor()
params={'n_estimators':[100,200,300,400,500,600,700,800,900,1000]}

grid_model = GridSearchCV(rfc,params,verbose=2)
grid_model.fit(X_train,y_train)

pred=grid_model.predict(X_test)

print('Random Forest accuracy is -->',r2_score(y_test,pred)*100)
```

```python
grid_model.best_params_
```

```python
res=pd.DataFrame()
res['Y-test']=y_test
res['pred']=pred
res.head()
```
